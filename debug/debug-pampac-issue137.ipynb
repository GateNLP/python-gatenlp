```python
import regex
from gatenlp import Document
from gatenlp.pam.pampac import *
from gatenlp.lib_spacy import apply_spacy
import spacy


```

```python
nlp = spacy.load("en_core_web_sm")
doc=Document ("The wheel of the car was found near the restaurant of the gas station of my friend 8 ")
doc=apply_spacy(nlp,doc,  setname="Original markups")
doc
```

```python
outset = doc.annset("debug1")
outset.clear()
Pampac(
    Rule(
        Seq(AnnAt("NounChunk"),
            N(
                Seq(
                    AnnAt("Token"),
                    AnnAt("NounChunk")
                ),
                min=1, max=2
            ),
            name="chunk"
        ),
        AddAnn(name="chunk", type="FOUND")
    )
).run(doc, doc.annset("Original markups").with_type("NounChunk", "Token"), outset=outset)
doc
```

```python
outset = doc.annset("debug2")
outset.clear()
Pampac(
    Rule(
        Seq(AnnAt("NounChunk"),
            N(
                Seq(
                    AnnAt("Token"),
                    AnnAt("NounChunk")
                ),
                min=1, max=2
            ),
            name="chunk"
        ),
        AddAnn(name="chunk", type="FOUND")
    ),
    skip = "one"
).run(
    doc, 
    doc.annset("Original markups").with_type("NounChunk", "Token"), 
    outset=outset
)
for ann in outset:
    print(ann)
doc
```
